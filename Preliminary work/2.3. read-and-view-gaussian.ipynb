```python
import torch


# Load the Gaussian representation
W= torch.load('/kaggle/input/mnist-gaussian-1/mnist_gaussian_representation.pt')
print(W.shape)
```

```python
# Assuming W is already defined and contains the Gaussian parameters

# Apply activation functions to get actual parameter values
sigma_x_values = torch.sigmoid(W[:, 0])  # Apply sigmoid to sigma_x
sigma_y_values = torch.sigmoid(W[:, 1])  # Apply sigmoid to sigma_y
rho_values = torch.tanh(W[:, 2])         # Apply tanh to rho
alpha_values = torch.sigmoid(W[:, 3])    # Apply sigmoid to alpha
colour_values = torch.clamp(W[:, 4], 0, 1)  # Grayscale intensity, clamped between 0 and 1
pixel_coords = W[:, 5:7]                 # Coordinates are already in range [-1, 1]

# Iterate over each Gaussian and print the parameters
for i in range(W.shape[0]):
    sigma_x = sigma_x_values[i].item()
    sigma_y = sigma_y_values[i].item()
    rho = rho_values[i].item()
    alpha = alpha_values[i].item()
    colour = colour_values[i].item()
    coord_x = pixel_coords[i, 0].item()
    coord_y = pixel_coords[i, 1].item()
    
    print(f"Gaussian {i+1}:")
    print(f"  Sigma_x: {sigma_x:.4f}")
    print(f"  Sigma_y: {sigma_y:.4f}")
    print(f"  Rho: {rho:.4f}")
    print(f"  Alpha: {alpha:.4f}")
    print(f"  Colour (Intensity): {colour:.4f}")
    print(f"  Coordinates: ({coord_x:.4f}, {coord_y:.4f})\n")
```
